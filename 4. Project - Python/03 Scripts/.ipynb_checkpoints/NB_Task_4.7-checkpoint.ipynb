### Direction 4

This one is a bit longer. All due to the fact, that I thought it would be cool, to have this process kinda automated, meaning that I wrote a script that can be run everytime changes occur and the time_periods and labels automatically work.

So the idea is, to first establish a way of automating the chosing of what are busy hours and what are less busy hours. I decided to use the IQRs which i get from the describe() function. Then I labelled the hours using the IQRs. And at the end merge this back into the main df.

```python
ords_prods_merged_new['order_hour_of_day'].value_counts(dropna = False)
```

```python
# This code takes the count of the hours and transforms it into a new df. Then it saves the describe function of this df
# and saves it under a nwe dataframe. Then It takes the values for the 25% and 75% quartile and saves those under 
# two variables. And finally prints those to check if it worked.

df_hour_count = ords_prods_merged_new['order_hour_of_day'].value_counts(dropna = False)

df_hour_described = df_hour_count.describe()

first_q = df_hour_described.values[4]

fourth_q = df_hour_described.values[6]

print(first_q)
print(fourth_q)
```

```python
# Here I take the df with the count of the hours and reset it's index under a new df

df_hour_count_r = df_hour_count.reset_index()
```

```python
# Here I rename the columns of the df with the resetted index

df_hour_count_r = df_hour_count_r.rename(columns={"index" : "hour_of_day", "order_hour_of_day" : "count"})
```

```python
# Here I used a function, since the df is so small
#
# This function adds a label if the count is either between, higher or lower than the 25% / 75% quartile.

def label_hours(row):
    if row['count'] <= first_q:
        return 'fewest orders'
    elif (row['count'] > first_q) and (row['count'] <= fourth_q):
        return 'average orders'
    elif row['count'] > fourth_q:
        return 'most orders'
    else: return 'not enough data'
```

```python
# The function is applied to the df

df_hour_count_r['labels'] = df_hour_count_r.apply(label_hours, axis=1)
```

```python
# Here I check the DF if it worked properly

df_hour_count_r
```

```python
amount_most_orders = (df_hour_count_r['labels'] == 'most orders').sum()

amount_average_orders = (df_hour_count_r['labels'] == 'average orders').sum()

amount_fewest_orders = (df_hour_count_r['labels'] == 'fewest orders').sum()

print(amount_most_orders)
print(amount_average_orders)
print(amount_fewest_orders)
```

```python
list_most_orders = []
list_average_orders = []
list_fewest_orders = []


for i in range(0, amount_most_orders):
    a = df_hour_count_r.loc[df_hour_count_r['labels'] == 'most orders']['hour_of_day'].values[i]
    list_most_orders.append(a)


for ii in range(0, amount_average_orders):
    aa = df_hour_count_r.loc[df_hour_count_r['labels'] == 'average orders']['hour_of_day'].values[ii]
    list_average_orders.append(aa)
    
for iii in range(0, amount_fewest_orders):
    aaa = df_hour_count_r.loc[df_hour_count_r['labels'] == 'fewest orders']['hour_of_day'].values[iii]
    list_fewest_orders.append(aaa)



list_most_orders.sort()
list_average_orders.sort()
list_fewest_orders.sort()


#print(list_most_orders)
#print(list_average_orders)
#print(list_fewest_orders)



    
```

### Sorting the lists into time-periods

The following 4 codefragments have the simple function of taking the lists and transforming them into time periods, so out of [0, 1, 2, 3, 4, 5] will become ['0-5']

The idea is the following, i go through the list and test if the second result minus the first result is '-1' or to be clear is NOT '-1'. If that's the case I will save the distance between the first and the last number that are on this list so far. And save this as a string into a new list. If I ran through the list from the for-loop I finish the code and save the last distance as well as a string.

The following is the test version I created to check if my work would withstand change in the dataframe

```python
# THE TEST VERSION

test_list = [1, 2, 3, 4, 5, 9, 10, 11, 14, 15, 16, 18, 19, 20]

#print(test_list)

ende = 13

x = 0
y = 1

list_start = 0
list_ende = 1

list_interval = []

for z in range(0, ende):
    
    w = test_list[x] - test_list[y]
    
    x = x + 1
    
    y = y + 1
    
    if w != -1:

        int1 = str(test_list[list_start])
        
               
        int2 = str(test_list[list_ende-1])
        
        int3 = int1 + '-' + int2
        
        list_interval.append(int3)            
            
        list_start = list_ende
        
    
    if z == ende-1:
        
        list_start = test_list[list_start]
        
        int1 = str(list_start)
             
        int2 = str(test_list[list_ende])
        
        int3 = int1 + '-' + int2
        
        list_interval.append(int3)        

    list_ende = list_ende + 1
    

print(list_interval)   
```

```python
# THE VERSION FOR THE MOST ORDERS

mostx = 0
mosty = 1

most_start = 0
most_ende = 1

list_most_interval = []

for mostz in range(1, amount_most_orders):
    
    mostw = list_most_orders[mostx] - list_most_orders[mosty]
    
    mostx = mostx + 1
    
    mosty = mosty + 1
    
    if mostw != -1:
        
        mostint1 = str(list_most_orders[most_start])
        
        mostint2 = str(list_most_orders[most_ende-1])
        
        mostint3 = mostint1 + '-' + mostint2
        
        list_most_interval.append(mostint3)
        
        most_start = most_ende
        
    if mostz == amount_most_orders-1:
        
        most_start = list_most_orders[most_start]
        
        mostint1 = str(most_start)
        
        mostint2 = str(list_most_orders[most_ende])
        
        mostint3 = mostint1 + '-' + mostint2
        
        list_most_interval.append(mostint3)
        
    most_ende = most_ende + 1
    
print(list_most_interval)
```

```python
# THE VERSION FOR THE AVERAGE ORDERS

averagex = 0
averagey = 1

average_start = 0
average_ende = 1

list_average_interval = []

for averagez in range(1, amount_average_orders):
    
    averagew = list_average_orders[averagex] - list_average_orders[averagey]
    
    averagex = averagex + 1
    
    averagey = averagey + 1
    
    if averagew != -1:
        
        averageint1 = str(list_average_orders[average_start])
        
        averageint2 = str(list_average_orders[average_ende-1])
        
        averageint3 = averageint1 + '-' + averageint2
        
        list_average_interval.append(averageint3)
        
        average_start = average_ende
        
    if averagez == amount_average_orders-1:
        
        average_start = list_average_orders[average_start]
        
        averageint1 = str(average_start)
        
        averageint2 = str(list_average_orders[average_ende])
        
        averageint3 = averageint1 + '-' + averageint2
        
        list_average_interval.append(averageint3)
        
    average_ende = average_ende + 1
    
print(list_average_interval)
```

```python
# THE VERSION FOR THE FEWEST ORDERS

fewestx = 0
fewesty = 1

fewest_start = 0
fewest_ende = 1

list_fewest_interval = []

for fewestz in range(1, amount_fewest_orders):
    
    fewestw = list_fewest_orders[fewestx] - list_fewest_orders[fewesty]
    
    fewestx = fewestx + 1
    
    fewesty = fewesty + 1
    
    if fewestw != -1:
        
        fewestint1 = str(list_fewest_orders[fewest_start])
        
        fewestint2 = str(list_fewest_orders[fewest_ende-1])
        
        fewestint3 = fewestint1 + '-' + fewestint2
        
        list_fewest_interval.append(fewestint3)
        
        fewest_start = fewest_ende
        
    if fewestz == amount_fewest_orders-1:
        
        fewest_start = list_fewest_orders[fewest_start]
        
        fewestint1 = str(fewest_start)
        
        fewestint2 = str(list_fewest_orders[fewest_ende])
        
        fewestint3 = fewestint1 + '-' + fewestint2
        
        list_fewest_interval.append(fewestint3)
        
    fewest_ende = fewest_ende + 1
    
print(list_fewest_interval)
```

```python
# JOINING THE LISTS AS A STRING

fewest_time_period = ', '.join(list_fewest_interval)

average_time_period = ', '.join(list_average_interval)

most_time_period = ', '.join(list_most_interval)


print(fewest_time_period)
print(average_time_period)
print(most_time_period)

```
