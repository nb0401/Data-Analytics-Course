##### Checking consistency

```python
# Data Type issues

for col in cust_df_dropped.columns.tolist():
    weird = (cust_df_dropped[[col]].applymap(type) != cust_df_dropped[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (cust_df_dropped[weird]) > 0:
        print (col)
```
