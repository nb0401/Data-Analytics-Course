#  4.6 Merging and exporting data

### This script contains the following points:

#### 1. Create data to experiment on
#### 2. Concatenate dataframes
#### 3. Append data
#### 4. Merge data


```python
# Importing Libraries

import pandas as pd
import numpy as np
import os

```

## 1. Create data to experiment on

```python
# Define a dictionary containing January 2020 data

data1 = {'customer_id':['6732', '767', '890', '635'], 'month':['Jan-20', 'Jan-20', 'Jan-20', 'Jan-20'], 
        'purchased_meat':[0, 13, 3, 4],
        'purchased_alcohol':[1, 2, 10, 0],
        'purchased_snacks':[10, 5, 1, 7]}
```

```python
# Define a dictionary containing February 2020 data

data2 = {'customer_id':['6732', '767', '890', '635'], 'month':['Feb-20', 'Feb-20', 'Feb-20', 'Feb-20'],
         'purchased_meat':[0, 10, 5, 3],
         'purchased_alcohol':[2, 4, 14, 0],
         'purchased_snacks':[15, 32, 2, 6]}
```

```python
# Define a third dicionary containing differnt columns

data3 = {'customer_id':['6732', '767', '890', '635'], 'month':['Jan-20', 'Jan-20', 'Jan-20', 'Jan-20'],
         'days_purchased_on':[0, 13, 3, 4]}
```

```python
# Convert the dictionary into a dataframe

df = pd.DataFrame(data1,index=[0, 1, 2, 3])
df_1 = pd.DataFrame(data2,index=[0, 1, 2, 3])

df_2 = pd.DataFrame(data3,index=[0, 1, 2, 3])
```

```python
df
```

```python
df_1
```

## 2. Concatenate dataframes

```python
# Concatenating the Dataframes

frames = [df, df_1]

df_concat = pd.concat(frames)

df_concat
```

## 3. Append data

```python
df_appended = df.append(df_1)

df_appended
```

## 4. Merge Data

```python
df_merged = df.merge(df_2, on = 'customer_id')

df_merged
```

```python
df_merged = df.merge(df_2, on = ['customer_id', 'month'])

df_merged
```

```python
df_merged = df.merge(df_2, on = ['customer_id', 'month'], indicator = True)

df_merged
```

```python
pd.merge(df, df_2, on = ['customer_id', 'month'], indicator = True)
```
