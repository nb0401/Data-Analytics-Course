#### Direction 7

```python
# Creating the order frequency flag
```

```python
# Again I don't know why, I'm getting this warning, but I'll just ignore it - Although, I guess, it comes from using the
# subset and not the original df.

subsetthistwo['median_orders'] = subsetthistwo.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
subsetthistwo.head(50)
```

```python
subsetthistwo.loc[subsetthistwo['median_orders'] > 20, 'frequency_customer'] = 'Non-frequent customer'
```

```python
subsetthistwo.loc[subsetthistwo['median_orders'] <= 10, 'frequency_customer'] = 'Frequent customer'
```

```python
subsetthistwo.loc[(subsetthistwo['median_orders'] > 10) & (subsetthistwo['median_orders'] <= 20), 'frequency_customer'] = 'Regular customer'
```
