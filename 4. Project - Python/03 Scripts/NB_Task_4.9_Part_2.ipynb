## 1. Importing Libraries and settings

```python
# Importing libraries

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy

plt.rcParams['figure.figsize']=(12.8, 7.2) # Increases the chart size to 1280x720 px
```

## 2. Importing Dataframe

```python
path = r'X:\Data_Analytic\JUPYTER\Instacart Basket Analysis'

df_full = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'full_merged_4-9-p1.pkl'))
```

## 3. Working on directions

#### Direction 2

```python
pd.set_option('display.max_columns', None)

df_full.head()
```

```python
# Creating a bar chart

bar = df_full['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
# Creating a histogram

histo = df_full['prices'].plot.hist(bins = 70)
```

```python
# Creating a Scatterplot

scatter = sns.scatterplot(x = 'prices', y = 'prices', data = df_full)
```

```python
# Creating a line chart with a sample subset

#Creating a sample

np.random.seed(4)

dev = np.random.rand(len(df_full)) <= 0.7

df_full_big = df_full[dev]

df_full_small = df_full[~dev]
```

```python
# Reducing columns

df_full_small_reduced = df_full_small[['orders_day_of_week', 'prices']]
```

```python
# Creating Line Chart

line = sns.lineplot(data = df_full_small_reduced, x = 'orders_day_of_week', y = 'prices')
```

#### Direction 6

```python
df_6 = df_full[['age', 'n_dependants']]

line_df_6 = sns.lineplot(data = df_6, x = 'age', y = 'n_dependants')

plt.xlabel('age', labelpad=5)
plt.ylabel('average number of dependants', labelpad=5)
plt.ticklabel_format(style='plain', axis='y')
plt.title('average number of dependants per age')
plt.grid()
```

#### Direction 7

```python
df_7 = df_full[['age', 'income']]

scatter_df_7 = sns.scatterplot(x = 'age', y = 'income', data = df_7)

plt.xlabel('age', labelpad=5)
plt.ylabel('income', labelpad=5)
plt.ticklabel_format(style='plain', axis='y')
plt.title('distribution of income per age')
plt.grid()
```

```python
scatter_df_7_2 = sns.scatterplot(x = 'income', y = 'age', data = df_7)
```

As we can see (a little bit better in the second scatterplot) the age range 18-41 has a max spending power of around 200000-400000 (the least amount of people), but the main area (largest amount of people) is around 25000-100000. Additionally at 100000 to 200000 there seems to be a mediocre amount of people.

The same is present at the age above 41. The largest amount of people has an income of around 25000 to 175000, then the mediocre amount is from 175000 to 300000 and the least amount of people is present at the income range from 300000 to 600000.

The lineplot below strengthens this observation of an increased spending power at above 41.

```python
line_df_7 = sns.lineplot(data = df_7, x = 'age', y = 'income')
```
