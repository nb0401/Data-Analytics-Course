## Explaining the profiling variable

This profiling variable uses the status of the orders regarding region (4), income (4), age (4), number of dependants (2), baby (2) and pet (2) to divide all counts of orders into 512 possible profiles. However, only 509 of those profiles exist. Therefore there are 3 sorts of customer profiles which haven't ordered yet.

The usefulness of this profiling variable is discussable, especially for charts, i would need at least 509 bars to show all customers.
